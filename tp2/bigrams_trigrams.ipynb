```python
import gensim
import spacy
from gensim import corpora
from gensim import models


#print(gensim.__version__)

# Lire le fichier txt
with open("exemple2.txt", "r", encoding="utf-8") as file:
    text = file.read()

text = text.replace('\n', ' ')

""" documents = [u"Football club Arsenal defeat local rivals this weekend.",
u"Weekend football frenzy takes over London.", u"Bank open for takeover bids after losing millions.", u"London football clubs bid to move to Wembley stadium.", u"Arsenal bid 50 million pounds for striker Kane.",
u"Financial troubles result in loss of millions for bank.", u"Western bank files for bankruptcy after financial losses.",
u"London football club is taken over by oil millionaire from Russia.", u"Banking on finances not working for Russia."] """

nlp = spacy.load("en_core_web_sm")
doc = nlp(text)
texts = []
for sentence in doc.sents :
    tokens = []
    for word in sentence:
         if not word.is_stop and not word.is_punct and not word.like_num :
            tokens.append(word.lemma_)
    texts.append(tokens)
    
""" for document in documents :
    text = []
    doc = nlp(document)
    for word in doc :
        if not word.is_stop and not word.is_punct and not word.like_num :
            text.append(word.lemma_)
    texts.append(text)
print (texts) """
```

```python
dictionary = corpora.Dictionary(texts)
print(dictionary.token2id)
```

```python
corpus = [ dictionary.doc2bow(text) for text in texts ]
""" corpus_list = [[(dictionary[id], freq) for id, freq in doc] for doc in corpus]

print(corpus_list) """

print (corpus)
#indice du token , frequence dans le document
```

```python
for i , doc in enumerate(corpus):
    print (f"document {i+1} : ,{doc}")
```

- To summarize, TF-IDF assigns higher weights to terms that are frequent in a particular document but infrequent across the entire corpus


```python

tfidf = models.TfidfModel(corpus)
for document in tfidf[corpus]:
    print (document)
```

* min_count : Ce paramètre contrôle combien de fois une paire de mots doit apparaître dans le corpus avant qu'elle ne soit considérée comme un bigram. Par défaut, ce paramètre est souvent plus élevé que 1, ce qui signifie que seules les paires apparaissant fréquemment seront détectées.

- Pourquoi l'ajouter ? En utilisant min_count=1, tu forces le modèle à détecter et former un bigram même s'il n'apparaît qu'une seule fois dans le corpus. Cela peut être utile si tu as un petit corpus ou si tu veux capturer des relations plus rares.
- threshold : Ce paramètre contrôle la sensibilité à laquelle les paires de mots seront considérées comme des bigrams. Un seuil plus bas (comme 2) rendra le modèle plus enclin à former des bigrams. Un seuil plus élevé rend le modèle plus conservateur.

bigram = gensim.models.Phrases(texts, min_count=1, threshold=2)

```python
#bigram = gensim.models.Phrases(texts, min_count= 3,threshold = 2)
bigram = gensim.models.Phrases(texts)
textsB = [bigram[line] for line in texts ]
#trigram = gensim.models.Phrases(textsB, min_count= 2,threshold = 2)
trigram = gensim.models.Phrases(textsB)
textsT = [trigram[line] for line in textsB ]
```

```python
for i,line in enumerate(textsB):
    print(f"Document {i+1} avec bigrams : {line}")
```

```python
for i,line in enumerate(textsB):
    bigrams =[token for token in line if '_' in token]
    if bigrams :
        print(f"Document {i+1} avec bigrams : {bigrams}")
```

```python
for i,line in enumerate(textsT):
    trigrams = [token for token in line if '_' in token and len(token.split('_')) == 3] 
    if trigrams :
        print(f"Document {i+1} avec trigrams : {trigrams}")
```
